## Pandas Series -> 1D 

```python
import numpy as np
import pandas as pd
```

###  Series from list

```python

runs = [12, 14, 13, 14]
print(pd.Series(runs))

# custom index with name
marks = [67, 57, 89, 100]
subjects = ['math', 'english', 'science', 'hindi']
print(pd.Series(marks, index=subjects, name='Sumaiyar marks'))
```

### Series from dictionary

```python


marks = {
  'math': 67,
  'english': 57,
  'science': 89,
  'hindi': 100
}

marks_series = pd.Series(marks, name='Sumaiyar marks')
print(marks_series)
```

```python
print(marks_series.size)
print(marks_series.dtype)
print(marks_series.name)
print(marks_series.is_unique)   # series ta unique kina

print(marks_series.index)   #shb index dibe
print(marks_series.values)
```

### Series from csv file dataset

```python
# with 1 col
subs = pd.read_csv('subs.csv')
subs = subs.squeeze('columns')
print(type(subs))
print(subs)     # col er name, store hoye gece name er vitor
```

```python
# with 2 col

vk = pd.read_csv('kohli_ipl.csv', index_col='match_no')
vk = vk.squeeze('columns')    # dataframe to series e convert korlo
print(vk)

# with 3 col  -> ekhane index movie and values lead actor
movies = pd.read_csv('bollywood.csv', index_col='movie')
movies = movies.squeeze('columns')
print(movies)
```

### Series with python functionality

```python
print(len(subs))
print(type(subs))
print(dir(subs))
print(sorted(subs))   # sort kore and output list e store kore
print(min(subs))
print(max(subs))
```

```python
# type conversion
print(list(subs))
print(dict(subs))

# membership operator
print('2 States (2014 film)' in movies)   # 133 subs e ace kina index e
print('Alia Bhatt' in movies.values)
```

```python
# looping
# for i in movies:
#   print(i)
```

```python
# Arithmetic operator -> broadcasting
print(500-subs)

# Relational operator
print(100 < subs)
```

```python
# find no of >= 50 scored by kohli
print(vk[vk >= 50].size)
# find no of ducks
print(vk[vk == 0].size)

# find days subs > 200
print(subs[subs > 200].size)

# kon actor 20 er beshi movie banaice
# print(movies.value_counts())  # ke koto movie banaice
num_movies = movies.value_counts()
# print(num_movies)
print(num_movies[num_movies > 20])
print(num_movies[num_movies > 20].size)
```

```python
# plot graph
subs.plot()
vk.plot()
```
