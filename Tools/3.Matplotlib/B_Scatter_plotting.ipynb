```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

#### Scatter Plots -> Features
- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
x = np.linspace(-10, 10, 50)
y = 10*x + np.random.randint(0, 300, 50)
y
```

```python
plt.scatter(x, y, color='red', marker='+', linewidths=2, )
plt.title('X and Y value')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# dataframe from pandas
df = pd.read_csv('datasets/batter.csv')
plt.scatter(df['avg'], df['strike_rate'], color='red', marker='+')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# scatter Vs plot
tips = sns.load_dataset('tips')

# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

### Colored Scatterplots

```python
iris = pd.read_csv('datasets/iris.csv')
iris.sample(5)
```

```python
# c = numeric => so Species ke agei numeric e convert kore nilam
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```
