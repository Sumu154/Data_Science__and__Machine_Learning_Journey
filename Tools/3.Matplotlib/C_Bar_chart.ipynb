## Bar Chart
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors, children, color='black')
# plt.barh(colors, children, color='black')  # horizontal bar er jonne barh
```

```python
# dataframe form pandas
df = pd.read_csv('datasets/batsman_season_record.csv')

plt.bar(df['batsman'], df['2015'])
```

```python
plt.bar(np.arange(df.shape[0])-0.25, df['2015'], width=0.25)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.25)
plt.bar(np.arange(df.shape[0])+0.25, df['2017'], width=0.25)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.xticks(rotation=30)
```

```python
# stacked bare chart
plt.bar(df['batsman'], df['2017'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom=df['2017'], label='2016')
plt.bar(df['batsman'], df['2015'], bottom=(df['2016']+df['2017']), label='2015')

plt.legend()
plt.show()
```
