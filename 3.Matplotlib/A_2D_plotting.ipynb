## Matplotlib plotting


#### Types of Data
- Numerical Data
- Categorical Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

#### 2D Line plot -> Features
- Bivariate Analysis
- Categorical -> numerical and numerical -> numerical
- Use case -> Time series data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year, price)
```

```python
# colors(hex) and line(width and style) and marker(size)
price = [48000,54000,57000,49000,47000,45000]
hype = [10000,20000,35000,49000,33000,35000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year, price, color='green', linestyle='solid', linewidth=3, marker='.', markersize=15, label='Car price')
plt.plot(year, hype, color='blue', linestyle='solid', linewidth=3, marker='.', markersize=15, label='Car hype')

plt.title('Year Vs Price')
plt.xlabel('Year')
plt.ylabel('price')

# plt.xlim(2017,2019)  # explicit limit thik kore dea jay
# plt.ylim(0,75000)

plt.legend()

plt.grid()
```
